# Importando dados

```python
import pandas as pd
```

```python
dados_2010_01 = pd.read_csv('.\Dados\emprestimos-20101.csv')
dados_2010_02 = pd.read_csv('.\Dados\emprestimos-20102.csv')
dados_2011_01 = pd.read_csv('.\Dados\emprestimos-20111.csv')
dados_2011_02 = pd.read_csv('.\Dados\emprestimos-20112.csv')
dados_2012_01 = pd.read_csv('.\Dados\emprestimos-20121.csv')
dados_2012_02 = pd.read_csv('.\Dados\emprestimos-20122.csv')
dados_2013_01 = pd.read_csv('.\Dados\emprestimos-20131.csv')
dados_2013_02 = pd.read_csv('.\Dados\emprestimos-20132.csv')
dados_2014_01 = pd.read_csv('.\Dados\emprestimos-20141.csv')
dados_2014_02 = pd.read_csv('.\Dados\emprestimos-20142.csv')
dados_2015_01 = pd.read_csv('.\Dados\emprestimos-20151.csv')
dados_2015_02 = pd.read_csv('.\Dados\emprestimos-20152.csv')
dados_2016_01 = pd.read_csv('.\Dados\emprestimos-20161.csv')
dados_2016_02 = pd.read_csv('.\Dados\emprestimos-20162.csv')
dados_2017_01 = pd.read_csv('.\Dados\emprestimos-20171.csv')
dados_2017_02 = pd.read_csv('.\Dados\emprestimos-20172.csv')
dados_2018_01 = pd.read_csv('.\Dados\emprestimos-20181.csv')
dados_2018_02 = pd.read_csv('.\Dados\emprestimos-20182.csv')
dados_2019_01 = pd.read_csv('.\Dados\emprestimos-20191.csv')
dados_2019_02 = pd.read_csv('.\Dados\emprestimos-20192.csv')
dados_2020_01 = pd.read_csv('.\Dados\emprestimos-20201.csv')
dados_exemplares = pd.read_parquet('.\Dados\dados_exemplares.parquet')
```

# Concatenando os dados em um DataFrame

```python
emprestimos = pd.concat(objs=[dados_2010_01, dados_2010_02, 
                              dados_2011_01, dados_2011_02, dados_2012_01, dados_2012_02, dados_2013_01, 
                              dados_2013_02, dados_2014_01, dados_2014_02, dados_2015_01, dados_2015_02, 
                              dados_2016_01, dados_2016_02, dados_2017_01, dados_2017_02, dados_2018_01, 
                              dados_2018_02, dados_2019_01, dados_2019_02, dados_2020_01])
emprestimos
```

## Mesclando os dados da tabela de empréstimo com a tabela de exemplares

```python
emprestimo_com_exemplares = pd.merge(emprestimos, dados_exemplares, how='left', on=None)
emprestimo_com_exemplares.head(10)
```
